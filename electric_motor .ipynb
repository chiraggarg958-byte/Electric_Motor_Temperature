```python
# --- IMPORTS ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import joblib

# --- BASIC SETUP ---
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
%matplotlib inline

# --- LOAD DATA ---
df = pd.read_csv(r"C:\Users\Asus\Desktop\measures_v2.csv")
print("‚úÖ Dataset Loaded Successfully!\n")

# --- DATA PREVIEW ---
display(df.head())
print("\n‚ÑπÔ∏è DataFrame Info:")
df.info()

print("\nüìä Statistical Summary:")
display(df.describe())

# --- CHECK PROFILE ID ---
if 'profile_id' in df.columns:
    plt.figure(figsize=(15,6))
    df['profile_id'].value_counts().sort_values().plot(kind='bar', color='skyblue')
    plt.title('Number of Measurements per Profile ID')
    plt.xlabel('Profile ID')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Column 'profile_id' not found ‚Äî skipping bar graph.")

# --- DISTRIBUTION PLOTS ---
print("\nüìà Plotting numeric distributions...")
for col in df.select_dtypes(exclude='object').columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df[col], kde=True, color='green')
    plt.axvline(df[col].mean(), color='red', linestyle='--', linewidth=2, label='Mean')
    plt.title(f'Distribution of {col}')
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- SKEWNESS ---
print("\nüìä Skewness of Numerical Features:")
display(df.skew(numeric_only=True))

# --- SCATTER PLOTS ---
print("\nüìâ Scatter plots with target 'pm'...")
if 'pm' in df.columns:
    fig, axes = plt.subplots(2, 4, figsize=(20, 6), sharey=True)
    features = ['ambient', 'coolant', 'motor_speed', 'i_d', 'u_q', 'u_d', 'i_q']
    for ax, feat in zip(axes.flat, features):
        if feat in df.columns:
            sns.scatterplot(x=feat, y='pm', data=df, ax=ax)
    fig.suptitle("Scatterplots: Parameters vs Rotor Temperature (pm)", fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Column 'pm' not found ‚Äî skipping scatter plots.")

# --- CORRELATION HEATMAP ---
print("\nüî• Generating Correlation Heatmap...")
num_df = df.select_dtypes(include='number').copy()
num_df.drop(columns=['profile_id'], inplace=True, errors='ignore')
plt.figure(figsize=(14,7))
sns.heatmap(num_df.corr(), annot=True, cmap='rocket', fmt=".2f")
plt.title("Correlation Heatmap", fontsize=15)
plt.tight_layout()
plt.show()

# --- PROFILE TEMPERATURE TREND ---
print("\nüìà Temperature trends for one profile...")
if 'profile_id' in df.columns:
    sample_id = 20
    if sample_id in df['profile_id'].unique():
        temp_data = df[df['profile_id'] == sample_id]
        plt.figure(figsize=(20,5))
        for col in ['stator_yoke', 'stator_tooth', 'stator_winding']:
            if col in df.columns:
                plt.plot(temp_data[col], label=col)
        plt.legend()
        plt.title(f"Temperature Comparison for Profile ID {sample_id}")
        plt.xlabel("Record Index (time sequence)")
        plt.ylabel("Temperature")
        plt.tight_layout()
        plt.show()
    else:
        print(f"‚ö†Ô∏è Profile ID {sample_id} not found.")
else:
    print("‚ö†Ô∏è 'profile_id' column not found ‚Äî skipping line plot.")

# --- DROP IDENTIFIERS & PREP DATA ---
print("\nüóëÔ∏è Cleaning and splitting data for model...")
target = 'pm'
drop_cols = [target, 'stator_yoke', 'stator_tooth', 'stator_winding', 'torque', 'profile_id']
X = df.drop(columns=[c for c in drop_cols if c in df.columns])
y = df[target].copy()

# --- TRAIN/TEST SPLIT ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=3)

# --- SCALING ---
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
X_train = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test  = pd.DataFrame(X_test_scaled,  columns=X_test.columns,  index=X_test.index)

# --- SAVE SCALER ---
joblib.dump(scaler, "transform.save")
print("\nüíæ Scaler saved as 'transform.save'")
print("‚úÖ Data processing and visualization completed successfully!")
```
